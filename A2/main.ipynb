## How to use:

```python
from PriceLoader import PriceLoader

# will intitalize and load all data
x = PriceLoader(start_date = '2024-01-01', end_date = '2025-01-01')

# fetch dataframe with data (* for all, ticker for specific e.g ticker = "AMD")
total_df = x.load_price(ticker = "*")

total_df['Close']
# total_df['Volume']
```

```python
long_df = x.long_format(total_df)
# replace NaN with 0s as requested
long_df = long_df.fillna(0)
long_df
```

```python
long_df.to_csv("data/market_data.csv", index=False)
```
